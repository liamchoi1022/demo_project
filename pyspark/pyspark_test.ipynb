```python
from pyspark.sql import SparkSession
user = "airflow"
password = "airflow"
```

```python
spark = SparkSession \
    .builder \
    .appName("pyspark demo_project") \
    .config("spark.jars", "postgresql-42.6.0.jar") \
    .getOrCreate()
```

```python
df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://localhost:5432/demo_project") \
    .option("dbtable", "silver.hourly_forecast") \
    .option("user", user) \
    .option("password", password) \
    .option("driver", "org.postgresql.Driver") \
    .load()
```

```python
df.show()
```
